```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import linear_model


house_data = pd.read_csv(r'D:\dataset\train.csv')

house_data
```

```python
(house_data.corr() > 0.4).sum() 
```

# For Univariate Linear Regression 

```python
from sklearn.model_selection import train_test_split


x = np.array(house_data['GrLivArea'])
y = np.array(house_data['SalePrice'])

x = x.reshape((len(x) , 1))
y = y.reshape((len(x) , 1))

training_data , testing_data , training_data_labels , testing_data_labels  = train_test_split( x , y , test_size = 0.2 , random_state = 5)

print("Trainig Data: " , training_data.shape)

print("training data labels: " , training_data_labels.shape)


print("Testing Data" , testing_data.shape)


print("Testing data labels: " , testing_data_labels.shape)
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()


model.fit(training_data , training_data_labels)
```

```python
plt.scatter(training_data , training_data_labels , color = 'orange' , marker = 'o' , label = 'Data Points')

plt.plot(training_data , model.predict(training_data) , color = 'green' , label = 'Regression Line')

plt.legend()
```

```python
result = model.predict(testing_data)

result
```

```python
from sklearn.metrics import accuracy_score , r2_score

r2 = r2_score(testing_data_labels, result)

r2
```

```python
accuracy = model.score(testing_data, testing_data_labels)

accuracy
```

```python
model.coef_
```

model.intercept_

## Polynomial Linear Regression

```python
from sklearn import linear_model

from sklearn.model_selection import train_test_split


from sklearn.preprocessing import PolynomialFeatures

house_data = pd.read_csv(r'D:\dataset\train.csv')

house_data

x = (house_data[['GrLivArea' , 'OverallQual' , '1stFlrSF' , 'GarageCars' , 'GarageArea' , 'TotalBsmtSF']])
y = (house_data['SalePrice'])

training_data , testing_data , training_data_labels , testing_data_labels  = train_test_split( x , y , test_size = 0.2 , random_state = 5)


poly_features = PolynomialFeatures(degree = 2)

x_poly = poly_features.fit_transform(training_data)

x_test_poly = poly_features.fit_transform(testing_data)



model = linear_model.LinearRegression()

model.fit(x_poly , training_data_labels)

y2_pred = model.predict(x_test_poly)
```

```python
y2_pred
```

```python
from sklearn.metrics import accuracy_score , r2_score

accuracy = r2_score( y2_pred , testing_data_labels)
accuracy
```
